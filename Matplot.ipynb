```python
import matplotlib.pyplot as plt
import numpy as np
"""
x = [1,2,3,4,5]
y = [10,15,13,18,16]

plt.plot(x, y)
plt.title("basic Line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
plt.plot(x,y, color='red', marker='o',linestyle='--', linewidth=2)
plt.show()

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [25,16,9,4,1]

plt.plot(x,y1, label= "Ascending")
plt.plot(x,y2, label= "Descending")
plt.legend()
plt.show()"""

#Bar charts
categories = ['A', 'B', 'C', 'D', 'E', 'F','G']
values = [10,20,36,18,20,50,46]
colors = ['violet','indigo','blue','green','yellow','orange','red']
plt.bar(categories,values, color=colors)
plt.title("Bar Chart Example")
plt.grid(True)
plt.show()

#Scatter Plot
x=[4,7,8,7,2,17,2,9]
y=[99,86,87,88,100,86,103,87]

plt.scatter(x,y, color='red')
plt.title("Scatter Plot Example")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()

#Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='teal', alpha=0.7,edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

#Pie Chart
sizes = [30,40,22,18]
labels = ['Python', 'java', 'C++', 'JavaScript']
colors = ['gold','skyblue','lightgreen','pink']
plt.pie(sizes, labels=labels,colors=colors, autopct='%1.1f%%', startangle=90,shadow=True,explode=[0.1,0,0,0])
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.axis('equal')
plt.show()

#Sub-Plots
x=[1,2,3,4,5]
y=[5,7,4,6,8]

plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.bar(x,y)
plt.tight_layout()
plt.show()
```
