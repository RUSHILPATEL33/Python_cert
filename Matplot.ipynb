```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5]
y = [10,15,13,18,16]

plt.plot(x, y)
plt.title("basic Line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
plt.plot(x,y, color='red', marker='o',linestyle='--', linewidth=2)
plt.show()

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [25,16,9,4,1]

plt.plot(x,y1, label= "Ascending")
plt.plot(x,y2, label= "Descending")
plt.legend()
plt.show()
```
